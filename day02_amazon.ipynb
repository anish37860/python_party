#### Q1. What is the average click-through rate (CTR) for sponsored product ads for each product category that contains the substring 'Electronics' in its name during October 2024? This analysis will help determine which electronics-related categories are performing optimally.


CTR = (Number of Clicks) / (Number of Impressions)

```python
df = dim_product.merge(fct_ad_performance, on='product_id', how='left')
df
```

```python
df['ctr'] = df['clicks']/df['impressions']
df
```

```python
df['recorded_date'] = pd.to_datetime(df['recorded_date'])

oct_data =df[ (df['recorded_date'].dt.year == 2024) 
                & (df['recorded_date'].dt.month == 10)]
oct_data
```

```python
elec_df = oct_data[ oct_data['product_category'].str.contains('Electronics') ]
elec_df
```

```python
avg_ctr_category = elec_df.groupby('product_category')['ctr']
avg_ctr_per_category = avg_ctr_category.mean()
avg_ctr_per_category
```

```python
avg_ctr_per_category = avg_ctr_per_category.reset_index()
avg_ctr_per_category
```

##### Submitted Code

```python
# Note: pandas and numpy are already imported as pd and np
# The following tables are loaded as pandas DataFrames with the same names: fct_ad_performance, dim_product
# Please print your final result or dataframe
df = dim_product.merge(fct_ad_performance, on='product_id', how='left')
df['ctr'] = df['clicks']/df['impressions']

df['recorded_date'] = pd.to_datetime(df['recorded_date'])

oct_data =df[ (df['recorded_date'].dt.year == 2024) 
                & (df['recorded_date'].dt.month == 10)]

elec_df = oct_data[ oct_data['product_category'].str.contains('Electronics') ]

avg_ctr_per_category = elec_df.groupby('product_category')['ctr']
avg_ctr_per_category = avg_ctr_per_category.mean()
avg_ctr_per_category = avg_ctr_per_category.reset_index()
print(avg_ctr_per_category)
```

#### Q2. Which product categories have a CTR greater than the aggregated overall average CTR for sponsored product ads during October 2024? This analysis will identify high-performing categories for further optimization. For this question, we want to calculate CTR for each ad, then get the average across ads by product category & overall.

```python
oct_data
```

```python
avg_ctr_per_category = oct_data.groupby('product_category')['ctr'].mean().reset_index()
avg_ctr_per_category
```

```python
overall_avg_ctr = oct_data['ctr'].mean()
overall_avg_ctr
```

```python
greater_than_avg_ctr = avg_ctr_per_category[avg_ctr_per_category['ctr'] > overall_avg_ctr]
greater_than_avg_ctr
```

##### Submitted Code 

```python
df = dim_product.merge(fct_ad_performance, on='product_id', how='left')
df['ctr'] = df['clicks']/df['impressions']

df['recorded_date'] = pd.to_datetime(df['recorded_date'])

oct_data =df[ (df['recorded_date'].dt.year == 2024) 
                & (df['recorded_date'].dt.month == 10)]

avg_ctr_per_category = oct_data.groupby('product_category')['ctr'].mean().reset_index()

overall_avg_ctr = oct_data['ctr'].mean()

greater_than_avg_ctr = avg_ctr_per_category[avg_ctr_per_category['ctr'] > overall_avg_ctr]
print(greater_than_avg_ctr)
```

#### Q3. For the product categories identified in the previous question, what is the percentage difference between their CTR and the overall average CTR for October 2024? This analysis will quantify the performance gap to recommend specific categories for targeted advertising optimization.

##### Submitted Code

```python
greater_than_avg_ctr['percentage_diff'] = ((greater_than_avg_ctr['ctr'] - overall_avg_ctr)/overall_avg_ctr)*100
greater_than_avg_ctr
```
